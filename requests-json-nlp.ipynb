3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

## Question 3 - Week 4 Project

```python
import json
import requests
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

def fetch_and_analyze_lyrics(band, song, filename):
    """
    Fetches lyrics from the lyrics.ovh API, saves them to a JSON file.

    Args:
        band (str): Name of the band or artist.
        song (str): Name of the song.
        filename (str): File path to save the lyrics JSON.

    Returns:
        dict: Dictionary with full lyrics text.
    """
    # Build API URL
    url = f"https://api.lyrics.ovh/v1/{band}/{song}"

    # Fetch lyrics
    response = requests.get(url)
    if response.status_code != 200:
        raise ValueError(f"Failed to fetch lyrics: {response.status_code} - {response.text}")

    result = json.loads(response.text)

    # Save to JSON file
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(result, f, indent=2, ensure_ascii=False)

    # Output result
    # print("Lyrics:\n", result)  # Print first 500 chars


    return {
        "lyrics": result

    }


fetch_and_analyze_lyrics("The Beatles", "Yesterday", "yesterday.json")
```

Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? ## Question 4 - Week 4 Project

```python
import json
import requests
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

def analyze_lyrics(filename):
    """
    Rruns sentiment analysis using spaCy and spaCyTextBlob.

    Args:
        filename (str): File path to load the lyrics JSON.

    Returns:
        dict: Dictionary with polarity, subjectivity, and full lyrics text.
    """
    
    # Load spaCy model
    nlp = spacy.load("en_core_web_sm")

    # Add sentiment component if not already added
    if "spacytextblob" not in nlp.pipe_names:
        nlp.add_pipe("spacytextblob", last=True)

    # Load the saved lyrics
    with open(filename, "r", encoding="utf-8") as f:
        data = json.load(f)

    lyrics = data.get("lyrics", "")

    # Analyze sentiment
    doc = nlp(lyrics)

    # Output result
    print("Lyrics:\n", lyrics[:500], "...")  # Print first 500 chars
    print("\nTitle:", filename)
    print("\nSentiment:")
    print("Polarity:", doc._.blob.polarity)
    print("Subjectivity:", doc._.blob.subjectivity)

    return {
        "lyrics": lyrics,
        "polarity": doc._.blob.polarity,
        "subjectivity": doc._.blob.subjectivity
    }


analyze_lyrics("blackbird.json")
analyze_lyrics("cometogether.json")
analyze_lyrics("something.json")
analyze_lyrics("yesterday.json")
```
